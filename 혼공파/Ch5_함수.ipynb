##키워드 매개변수

```python
#기본 매개변수가 가변 매개변수보다 앞에 올 때
#오류 남 -> 기본 매개변수는 가변 매개변수 앞에 써도 의미 없어

def print_n_times(n=2, *values):
  for i in range(n):
    for value in values:
      print(value)
    print()

print_n_times("안녕하세요", "즐거운", "프로그래밍")
```

```python
#가변 매개변수가 기본 매개변수보다 앞에 올 때
#가변 매개변수가 우선시 됨 -> 그럼 같이 쓸 수 없을까? '키워드 매개변수'

def print_n_times(*values, n=2):
  for i in range(n):
    for value in values:
      print(value)
    print()

print_n_times("안녕하세요", "즐거운", "프로그래밍", 3)
```

```python
#키워드 매개변수

def print_n_times(*values, n=1):
  for i in range(n):
    for value in values:
      print(value)
    print()

print_n_times("안녕하세요", "즐거운", "프로그래밍", n=3) #n=3이 키워드 매개변수
```

##기본적인 함수의 활용

```python
#범위 내부의 정수를 모두 더하는 함수

def sum_all(start, end):
  output = 0
  for i in range(start, end+1): #0, 1, 2, ..., 100
    output += i
  return output

sum_all(0, 100) #0, 1, 3, 6, ....
```

```python
#기본 매개변수와 키워드 매개변수를 활용해 범위의 정수를 더하는 함수

def sum_all(start=0, end=100, step=1):
  output = 0
  for i in range(start, end+1, step):
    output += i
  return output

print(sum_all(end=100))
```

##재귀함수

```python
#반복문으로 팩토리얼 구하기

def factorial(n):
  output = 1
  for i in range(1,n+1):
    output *= i
  return output

print(factorial(1))
print(factorial(5))
print(factorial(10))
```

```python
#재귀 함수로 팩토리얼 구하기
#재귀(recursion): 자기 자신을 호출하는 것

def factorial(n):
  if n == 0:
    return 1
  else:
    return n*factorial(n-1)

factorial(10)
```

##재귀함수의 문제

```python
#재귀 함수로 구현한 피보나치 수열(1)

def fibonacci(n):
  if n == 1:
    return 1
  if n == 2:
    return 2
  else:
    return fibonacci(n-2) + fibonacci(n-1)

print(fibonacci(35)) #시간이 좀 오래걸림...
```

```python
#재귀 함수로 구현한 피보나치 수열(2)

counter = 0

def fibonacci(n):
  print("fibonacci({})를 구합니다.".format(n))
  global counter
  counter += 1
  if n == 1:
    return 1
  if n == 2:
    return 2
  else:
    return fibonacci(n-2) + fibonacci(n-1)

fibonacci(10)
print("fibonacci(10) 계산에 활용된 덧셈 횟수는 {}입니다.".format(counter))
```

```python
#메모화

dictionary = {
    1: 1,
    2: 1
}

def fibonacci(n):
  if n in dictionary:
    return dictionary[n]
  else:
    output = fibonacci(n-1) + fibonacci(n-2)
    dictionary[n] = output
    return output

print(fibonacci(10))
```

##리스트 평탄화하는 재귀함수 만들기

```python
#리스트 평탄화: 중첩된 리스트가 있을 때 중첩을 모두 제거하고 풀어서 1차원 리스트로 만드는 것

def flatten(data):
  output = []
  for item in data:
    if type(item) == list:
      output += flatten(item) #재귀함수
    else:
      output.append(item)
  return output

example = [[1,2,3], [4,[5,6]], [7,8]]
print(flatten(example))
```

```python
#람다의 개념
#lambda 매개변수: 리턴값

power = lambda x: x*x
under_3 = lambda x: x<3

list_input_a = [1, 2, 3, 4, 5]

output_a = map(power, list_input_a)
print(list(output_a))

output_b = filter(under_3, list_input_a)
print(list(output_b))
```

##파일 처리

- 파일 열고 닫기

  - r: read 모드(읽기 모드)

```python
#파일 열고 닫기

file = open("basic.txt", "w")

file.write("Hello Python Programming")

file.close()
```

- 텍스트 읽기
  - 텍스트를 쓸 때에는 write()함수

```python
#read()함수로 텍스트 읽기

with open("basic.txt", "r") as file:
  contents = file.read()

print(contents)
```

```python
#랜덤하게 1000명의 키와 몸무게 만들기

import random

hanguls = list("가나다라마바사아자차카타파하")

with open("info.txt", "w") as file:
  for i in range(1000):
    name = random.choice(hanguls) + random.choice(hanguls)
    weight = random.randrange(40, 100)
    height = random.randrange(140, 200)
    file.write("{}, {}, {}\n".format(name, weight, height))
```
